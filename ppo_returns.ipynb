```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_returns = pickle.load(open('all_returns.pkl', 'rb'))
```

```python
mean_returns = [np.max(returns) for returns in all_returns]
```

```python
plt.plot(mean_returns)
```

```python
max_returns = []
max_r = -100
for returns in mean_returns:
    if returns > max_r:
        max_returns.append(returns)
        max_r = returns
    else:
        max_returns.append(max_r)
```

```python
plt.plot(max_returns)
plt.savefig('max_returns.jpg', dpi=1000)
```
